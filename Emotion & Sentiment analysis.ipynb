```python
def Emotion7newsfx(texts):
    response = natural_language_understanding.analyze(
    text=texts,
    features=Features(
        emotion = DocumentEmotionResults(),concepts = ConceptsResult(),sentiment = DocumentSentimentResults())).get_result()
    try:
        sadness_7news.append(
                {
                'Sadness':response.get('emotion', {}).get('document',{}).get('emotion',{}).get('sadness'),
                }
                )
    except ApiException as e:
        sadness_7news.append(
                {
                'Sadness':"append-",
                }
                )
        print(e)
    try:  
        fear_7news.append(
                {
                'Fear':response.get('emotion', {}).get('document',{}).get('emotion',{}).get('fear'),
                }
                )
    except ApiException as e:
        fear_7news.append(
                {
                'Fear':"append-",
                }
                )
        print(e)
    try:
        disgust_7news.append(
                {
                'Disgust':response.get('emotion', {}).get('document',{}).get('emotion',{}).get('disgust'),
                }
                )
    except ApiException as e:
        disgust_7news.append(
                {
                'Disgust':"append-",
                }
                )
        print(e)
    try:
        anger_7news.append(
                {
                'Anger':response.get('emotion', {}).get('document',{}).get('emotion',{}).get('anger'),
                }
                )
    except ApiException as e:
        anger_7news.append(
                {
                'Anger':"append-",
                }
                )
        print(e)
    try:
        joy_7news.append(
                {
                'Joy':response.get('emotion', {}).get('document',{}).get('emotion',{}).get('joy'),
                }
                )
    except ApiException as e:
        joy_7news.append(
                {
                'Joy':"append-",
                }
                )
        print(e)
    try:
        xx = response.get('concepts')
        y=xx[0].get('text')
        Concepts_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Top Relevance':y,
                }
                )
    except ApiException as e:
        Concepts_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Top Relevance':"append-",
                }
                )
        print(e)
    except IndexError as e:
        Concepts_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Top Relevance':"Index-",
                }
                )
        print(e)
    except TypeError as e:
        Concepts_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Top Relevance':"Type-",
                }
                )
        print(e)
    try:
        sentiments_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Sentiments':response.get('sentiment', {}).get('document',{}).get('score'),
                }
                )
    except ApiException as e:
        sentiments_7news.append(
                {
                #predict sentiments_7news through nrclex
                'Sentiments':"append-",
                }
                )
        print(e)
    
    return print("success in 7news emotion")
```
